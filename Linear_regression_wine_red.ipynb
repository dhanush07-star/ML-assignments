**divide data into input and output data**

```python
df.columns
```

```python
X=df.drop('quality',axis=1)
y=df['quality']
```

$Step-4$

**Divide data into train and test**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,  # Input data
                                                  y,  # output data
                                                  random_state=1234, # it select random samples
                                                  test_size=0.30)
```

```python
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(X_train,
       y_train)
```

$Step-6$

**Model predictions**

```python
# Model predictions happens X_test
y_predictions=LR.predict(X_test)
```

```python
y_predictions
```

$Step-7$

**Model evaluation**

```python
# RMSE
# MSE
# MAE
# R-square

from sklearn.metrics import r2_score,mean_squared_error
```

```python
R2=r2_score(y_test,y_predictions)
MSE=mean_squared_error(y_test,y_predictions)
RMSE=np.sqrt(MSE)
#accuracy_score(y_test,y_predictions) # it is a regression tech
print("R-sqaure:",R2)
print("MSE:",MSE)
print("RMSE:",RMSE)
```
